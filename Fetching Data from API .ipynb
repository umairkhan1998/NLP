### Fetching Data from Api

Importing Libraries

```python
import pandas as pd
import requests as req
import warnings
warnings.filterwarnings("ignore")
```

```python
respond = req.get("https://api.themoviedb.org/3/movie/top_rated?api_key=8265bd1679663a7ea12ac168da84d2e8&language=en-US&page=1")
```

```python
df = pd.DataFrame(respond.json()["results"])
```

```python
df
```

```python
for i in range(1,460):
    respond = req.get('https://api.themoviedb.org/3/movie/top_rated?api_key=8265bd1679663a7ea12ac168da84d2e8&language=en-US&page={}'.format(i))
    temp_df= pd.DataFrame(respond.json()["results"])[['original_title','overview','genre_ids']]
    df=df.append(temp_df,ignore_index=True)
```

```python
df.head(1)
```

```python
Data = df[['original_title','overview','genre_ids']]
```

```python
Data['genre_ids']
```

```python
res1=req.get('https://api.themoviedb.org/3/genre/movie/list?api_key=8265bd1679663a7ea12ac168da84d2e8&language=en-US')
```

```python
res1.json()
```

```python
df2= pd.DataFrame(res1.json()['genres'])
```

```python
df2
```

```python

genre_id_to_name = pd.Series(df2.name.values, index=df2.id).to_dict()


def replace_ids_with_names(id_list):
    return [genre_id_to_name.get(genre_id, "Unknown") for genre_id in id_list]

# Apply the function to each row in df2
Data['genre_names'] = Data['genre_ids'].apply(replace_ids_with_names)

print(Data)
```

```python
Data
```

```python
dataframe=Data.rename(columns={"original_title":'name',"overview":'description',"genre_ids":'G_id',"genre_names":'Genre'})
```
